```python
import pandas as pd
from azure.ai.ml import MLClient
from azure.identity import DefaultAzureCredential

ml_client = MLClient.from_config(credential=DefaultAzureCredential())
data_asset = ml_client.data.get("Diabetes_Dataset", version="1")

df = pd.read_csv(data_asset.path)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))

for name, group in df.groupby('diabetes'):
    plt.hist(group['age'], bins=10, alpha=0.7, label=f'diabetes={name}')

    
plt.title(f'Age distribution for diabetes={name}')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))

for name, group in df.groupby('diabetes'):
    plt.hist(group['HbA1c_level'], bins=10, alpha=0.7, label=f'diabetes={name}')

    
plt.title(f'HbA1c_level distribution for diabetes={name}')
plt.xlabel('HbA1c_level')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
